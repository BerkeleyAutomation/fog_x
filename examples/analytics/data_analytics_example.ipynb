```python
import polars as pl
from PIL import Image
import io
import numpy as np 

def deserialize_array(serialized_array):
    value = np.load(io.BytesIO(serialized_array)) #.astype(feature_spec.np_dtype)
    
    result = np.mean(value)
    
    memfile = io.BytesIO()
    np.save(memfile, result)
    return memfile.getvalue()


lazy_df = all_step_data.with_columns([
    pl.col("natural_language_embedding").map_elements(lambda x: deserialize_array(x)).alias("mean"),
])
```
